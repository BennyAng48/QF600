## 2. Briefly explain the economic significance and pricing implications of the intercept and slope coefficients.

#### Session 3 Page 6-7,18 

The slope coefficients provide an estimate of the portfolio's beta, the amount of exposure to systematic market risk the portfolio has that cannot be diversified away by holding more risky assets in the portfolio. The higher the beta, the more exposed the industry portfolio is to market risk. Accordingly, an investor would expect to earn a higher market risk premium as economic compensation for taking on this market risk. If an asset has negative beta, it has negative correlation with market return and investors would be willing to pay a premium to own such assets to provide insurance against a market downturn.

The intercept represents the pricing error relative to CAPM for individual assets or 'passive' portfolios. A positive intercept indicates that the CAPM predicted returns is below the actual returns of the asset and thus the asset is underpriced according to the CAPM model. The asset will lie above the SML. It is posible that there are other systematic factors impacting the asset returns that are not captured by the CAPM (not reflected inn market beta) such as size, value and momentum.

## 5. Briefly explain the economic significance and pricing implications of the SML.

#### Session 3 Page 9-10

The SML is a graphical representation of the CAPM formula. It provides a linear relationship relating the expected return of an asset with respect to its beta. Assets with higher beta have a greater exposure to systematic market risk and hence investors would demand a higher expected return to hold such assets. Assets that lie on the SML have the same Treynor ratio and provide the same risk premium to beta ratio. 

Assets that lie above the SML are underpriced and investors will buy the asset, causing prices to rise and expected returns to fall. On the other hand, assets that lie below the SML are perceived to be overpriced, and investors will sell the asset, causing prices to fall until expected returns rise to the level as predicted by the SML.

# Functions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Market Model
def MarketModel(data_portfolio, array_rf, data_market):
    #Market Model
    x = data_market - array_rf
    y = data_portfolio - array_rf
    
    c = len(y.columns)
    
    #Regression of excess return of portfolio on market
    model = [LinearRegression().fit(x, y.iloc[:,i]) for i in range(c)]
    r_sq = [model[i].score(x, y.iloc[:,i]) for i in range(c)]
    alpha = [model[i].intercept_ for i in range(c)]
    beta = [model[i].coef_[0] for i in range(c)]
    
    MM_table = pd.DataFrame({'Alpha': alpha, 'Beta': beta, 'R2': r_sq}, 
                              index = y.columns)
    
    return MM_table
```

```python
#CAPM
def CAPM(data_portfolio, array_rf, data_market, MM_table, x0, xt, num):
    # CAPM
    MM_table = MarketModel(data_portfolio, array_rf, data_market)  
    r_portfolio = data_portfolio.apply(np.mean, axis=0)
    r_market = data_market.apply(np.mean, axis=0)
    
    # Creating the CAPM table with portfolio returns and beta values
    CAPM_table = pd.DataFrame(r_portfolio, index=data_portfolio.columns, columns=['Mean Return'])
    CAPM_table['Beta'] = MM_table['Beta']
    
    # Concatenate market information to the CAPM table
    market_row = pd.DataFrame({'Mean Return': r_market.values, 'Beta': [1]}, index=['Market'])
    CAPM_table = pd.concat([CAPM_table, market_row])
    
    # Regression of mean return of portfolio & market on beta
    x2 = CAPM_table['Beta'].values.reshape((-1, 1))  # Independent variable (Beta)
    y2 = CAPM_table['Mean Return'].values            # Dependent variable (Mean Return)
    
    model2 = LinearRegression().fit(x2, y2)
    r_sq2 = model2.score(x2, y2)       # R-squared
    alpha2 = model2.intercept_         # Intercept (Alpha)
    beta2 = model2.coef_[0]            # Slope (Beta)
    
    # Plot of SML
    x_SML = np.linspace(x0, xt, num)
    y_SML = alpha2 + beta2 * x_SML
    
    plt.figure()
    plt.xlabel('Beta')
    plt.ylabel('Mean Return (%)')
    plt.plot(x_SML, y_SML, 'r-', label='SML')  # Security Market Line
    plt.legend(loc='upper left')
    plt.scatter(x2, y2, color='blue')  # Scatter plot of portfolio and market data
    plt.show()
    
    # Output results
    print('Alpha of CAPM: ', np.round(alpha2, 4))
    print('Beta of CAPM: ', np.round(beta2, 4))
    print('R² of CAPM: ', np.round(r_sq2, 4))
    
    return CAPM_table
```
